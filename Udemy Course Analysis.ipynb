# LETS VISUALIZE THE BELOW THINGS

1- FINDING THE MOST ENROLLED COURSE

2- TOP 10 POPULAR COURSES

3- COURSE HAVING HIGHEST NO. OF REVIEWS

4- YEAR WISE REVENUE BY UDEMY

```python
# FINDING THE MOST ENROLLED COURSE
df.groupby('course_title')['num_subscribers'].sum().sort_values(ascending=False).head(1)
```

```python
#TOP 10 POPULAR COURSES

top10 = df.sort_values(by='num_subscribers',ascending =False).head(10)
sns.barplot(y='course_title',x='num_subscribers',data=top10)
plt.xticks(rotation=45)
plt.grid('True')
```

```python
# to visualize in pie plot

Popular_courses=df.groupby('course_title')['num_subscribers'].sum().sort_values(ascending=False).head(10)

plt.pie(Popular_courses,normalize=True,autopct='%1.1f%%',labels=Popular_courses.index)
```

```python
# COURSE HAVING HIGHEST NO. OF REVIEWS

df.groupby('subject')['num_reviews'].sum().sort_values(ascending=False)
```

```python
most_reviewed = df.groupby('subject')['num_reviews'].sum().sort_values(ascending=False)
sns.barplot(y=most_reviewed.index, x=most_reviewed.values, data=df)
plt.xticks(rotation=45)
plt.grid('True')
```

```python
# is price affecting no. of subsribers
sns.scatterplot(x='price',y='num_subscribers',data=df)
```

```python
# We can see that no of subcribers are decresing for high priced coursed
```

```python
# Year wise earnings by udemy
df.groupby(df['published_timestamp'].dt.year)['price'].sum()
```

```python
result = df.groupby(df['published_timestamp'].dt.year)['price'].sum()
result.plot(kind='bar')

plt.xlabel('Year')
plt.ylabel('Total Price')
plt.title('Total Price by Year')

plt.show()
```
